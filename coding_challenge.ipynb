
**Scenario 7: Shopping Cart System**

You are creating a simple billing system where customers can enter the prices of items they
want to purchase. The program should keep accepting prices until the user decides to stop, then
apply appropriate discounts before generating the final bill.
Tasks:
* Continuously accept item prices using a while loop.
* Stop taking input when the user enters 0 (sentinel value).
*Keep track of the total number of items and the subtotal amount.
* Ensure no negative prices are allowed (ask again without counting as an item).
*  Apply discounts based on the subtotal:
○ 20% for totals ≥ 1000
○ 10% for totals ≥ 500
○ 5% for totals ≥ 200
○ 0% for totals < 200
* Display a clear bill summary including subtotal, discount amount, and final total.

```python
subtotal = 0
item_count = 0

while True:
    try:
        price = float(input("Enter item price (0 to stop): "))

        if price == 0:
            break

        if price < 0:
            print("Negative prices are not allowed. Try again.")
            continue

        subtotal += price
        item_count += 1

    except ValueError:
        print("Invalid input. Please enter a number.")

if subtotal >= 1000:
    discount_rate = 0.20
elif subtotal >= 500:
    discount_rate = 0.10
elif subtotal >= 200:
    discount_rate = 0.05
else:
    discount_rate = 0.00

discount_amount = subtotal * discount_rate
final_total = subtotal - discount_amount


print("\n----- BILL SUMMARY -----")
print(f"Total items: {item_count}")
print(f"Subtotal: ₹{subtotal:.2f}")
print(f"Discount ({int(discount_rate*100)}%): ₹{discount_amount:.2f}")
print(f"Final Total: ₹{final_total:.2f}")
```


**Scenario 8: Employee Hours Report**

You are analyzing student exam scores to generate performance reports.
Tasks:
1. Create a list of 10 random exam scores ranging from 0 to 100 using list comprehension.
2. Display each score with its index using the enumerate function.
3. Find all scores greater than 75 and display them.
4. Calculate and display the average score.

```python
import random


scores = [random.randint(0, 100) for _ in range(10)]

# each score with its index using enumerate
print("Exam Scores with Index:")
for index, score in enumerate(scores):
    print(f"Index {index}: Score {score}")

# Finding all scores greater than 75 and display them
high_scores = [score for score in scores if score > 75]
print("\nScores greater than 75:", high_scores)

# display the average score
average_score = sum(scores) / len(scores)
print(f"\nAverage Score: {average_score:.2f}")
```


**Scenario 9: Simple Calculator App**

A company wants to provide users with a simple calculator to perform basic mathematical
operations.
Tasks:
1. Create functions for addition, subtraction, multiplication, and division.
2. Ask the user to input two numbers and the operation they want to perform.
3. Call the appropriate function and display the result.
4. Use built-in functions like abs(), round(), and pow() within your solution.
5. Handle division by zero gracefully and inform the user.

```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error: Division by zero is not allowed."
    return a / b

try:
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
    operation = input("Choose operation (+, -, *, /): ")


    if operation == "+":
        result = add(num1, num2)
    elif operation == "-":
        result = subtract(num1, num2)
    elif operation == "*":
        result = multiply(num1, num2)
    elif operation == "/":
        result = divide(num1, num2)
    else:
        result = "Invalid operation."


    if isinstance(result, (int, float)):
        print(f"\nResult: {result}")
        print(f"Absolute Value: {abs(result)}")
        print(f"Rounded Value: {round(result, 2)}")
        print(f"Power (result^2): {pow(result, 2)}")
    else:
        print(result)

except ValueError:
    print("Invalid input. Please enter numeric values.")
```

**Scenario 10: Bank Account Manager**

A bank wants to track account balances and transactions for each customer in an
object-orientated way.
Tasks:
1. Create a class BankAccount with account_holder and balance attributes.
2. Implement methods for deposit(), withdraw(), and display_balance().
3. Use both local and global variables appropriately, such as a global transaction counter.
4. Demonstrate calling methods using both positional and keyword arguments.
5. Prevent withdrawals that would result in a negative balance.

```python
transaction_count = 0

class BankAccount:
    def __init__(self, account_holder, balance=0):
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        global transaction_count
        self.balance += amount
        transaction_count += 1
        print(f"Deposited ₹{amount}")

    def withdraw(self, amount):
        global transaction_count
        if amount > self.balance:
            print("Insufficient funds. Withdrawal denied.")
        else:
            self.balance -= amount
            transaction_count += 1
            print(f"Withdrew ₹{amount}")

    def display_balance(self):
        print(f"{self.account_holder}'s Balance: ₹{self.balance}")
        print(f"Total Transactions: {transaction_count}")


# Positional arguments
acc1 = BankAccount("Arun", 500)
acc1.deposit(200)
acc1.withdraw(100)
acc1.display_balance()

print("---")

# Keyword arguments
acc2 = BankAccount(account_holder="Meena", balance=1000)
acc2.deposit(amount=300)
acc2.withdraw(amount=1500)
acc2.display_balance()
```

**Scenario 11: Student Management System **

You are tasked with building a Student Management System that integrates multiple Python
concepts to help staff manage students' data, performance, and reporting.
Tasks:
1. Use dictionaries and lists to store and retrieve student information such as name, age,
and grades.
2. Allow users to add, delete, update, and search for students.
3. Track grades and categorize performance using functions and conditionals.
4. Use tuples or frozensets to store constant data like available classes.
5. Optionally implement file handling to save and load student data between program
runs.
6. Use exception handling to manage invalid inputs.
7. Implement sorting and filtering of students for reports.

```python
import json

students = []  # list to store student dictionaries
FILENAME = "students.json"

def add_student(name, age, grade):
    students.append({"name": name, "age": age, "grade": grade})

def delete_student(name):
    for s in students:
        if s["name"].lower() == name.lower():
            students.remove(s)
            return

def search_student(name):
    for s in students:
        if s["name"].lower() == name.lower():
            print(s)
            return
    print("Not found")

def report():
    for s in students:
        if s["grade"] >= 90:
            perf = "Excellent"
        elif s["grade"] >= 75:
            perf = "Good"
        elif s["grade"] >= 50:
            perf = "Average"
        else:
            perf = "Poor"
        print(f"{s['name']} - {perf}")

def save_data():
    with open(FILENAME, "w") as f:
        json.dump(students, f)
    print("Data saved.")

def load_data():
    global students
    try:
        with open(FILENAME, "r") as f:
            students = json.load(f)
        print("Data loaded.")
    except FileNotFoundError:
        print("No saved data found.")


load_data()

add_student("Arun", 16, 85)
add_student("Meena", 17, 92)
add_student("Ravi", 15, 45)

search_student("Meena")
delete_student("Arun")

print("\n--- Report ---")
report()

save_data()
```
